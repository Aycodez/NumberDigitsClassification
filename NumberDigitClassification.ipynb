```python
# Lets import the necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits
from sklearn.svm import SVC
import numpy as np

import matplotlib.pyplot as plt
```

```python
#Loading our images
data = load_digits()
```

```python
#Splitting our data into features and labels
X = data.data
y = data.target
```

```python
print('Images:', X.shape)
print('Labels:', y.shape)
```

```python
# Lets plot one of the images
print("Label: ", y[0])
plt.imshow(X[0].reshape(8,8), cmap=plt.cm.gray)
```

```python
#Lets split our data for training and testing
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=23)
```

```python
print('X_train shape:', X_train.shape)
print('X_test shape:', X_test.shape)
print('y_train shape:', y_train.shape)
print('y_test shape:', y_test.shape)
```

```python
# Lets build our model
classifier = SVC()
classifier.fit(X_train,y_train)
```

```python
#Lets check the accuracy of our model
print('Accuracy:', classifier.score(X_test,y_test))
```

```python
# Lets plot some images and check our predictions
fig, axes = plt.subplots(3, 4, figsize=(8,8))
for i , ax in enumerate(axes.flat):
    pred_image = classifier.predict(X_test[i].reshape(1, -1))
    ax.set_title(f'Actual label: {y_test[i]} \n Predicted label: {pred_image[0]}')
    ax.imshow(X_test[i].reshape(8,8))
    
plt.tight_layout(pad=0.2,w_pad=0.3)
```
